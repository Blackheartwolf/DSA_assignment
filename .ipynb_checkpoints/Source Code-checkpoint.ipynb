# Machine Learning On 2021 Covid19 


Steps to take to create a model.

    1. Import the data
    2. Clean the Data
    3. Split the Data into Training 80% /Test Sets 20%
    4. Create a model: Select an algorithm, which will be selected based on what we would like to achieve.
    5. Train the model
    6. Make Predictions 
    7. Evaluate and improve


```python
ml_data =  Covid19_2021.loc['Hardap'][['New Cases ','Active Cases', 'Deaths']]
```

```python
#Prepare the data
#This is our input set X
X = ml_data.drop(columns=['Deaths'])
#This is our output set Y
y = ml_data['Deaths']

X_train, X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)


model = DecisionTreeClassifier()
model.fit(X_train,y_train)

# Calculating Accuracy
predictions = model.predict(X_test)
score = accuracy_score(y_test,predictions)
score
```
