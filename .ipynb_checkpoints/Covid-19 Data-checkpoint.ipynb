```python
#Import the libraries
import pandas as pd
import numpy as np


import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="ticks", color_codes = True)
```

```python
#Path
p1 = '2021_DSA_Cases.xlsx'
p2 = '2020_DSA_Cases.xlsx'
```

```python
#2020 Data
#One way to import data from one excel with multiple sheets 
#Using pd.read_excel
#Raw_Khomas_2021 = pd.read_excel(r'2020_DSA_Cases.xlsx',sheet_name = 'Khomas')
#Raw_Ohangwena_2021 = pd.read_excel(r'2020_DSA_Cases.xlsx',sheet_name = 'Ohangwena')
#Continue with all the regions


## Now you can list all sheets in the file
#xls.sheet_names
#Reading all the sheets within the xls.
xls = pd.ExcelFile(p2)
Kho_df = pd.read_excel(xls, 'Khomas')
Oha_df = pd.read_excel(xls, 'Ohangwena')
Ero_df = pd.read_excel(xls, 'Erongo') 
Omu_df = pd.read_excel(xls, 'Omusati')
Kun_df = pd.read_excel(xls, 'Kunene')
Osha_df = pd.read_excel(xls, 'Oshana')
KaE_df = pd.read_excel(xls, 'Kavango East')
KaW_df = pd.read_excel(xls, 'Kavango West')
Oshi_df = pd.read_excel(xls, 'Oshikoto')
Khar_df = pd.read_excel(xls, 'Kharas')
Zamb_df = pd.read_excel(xls, 'Zambezi')
Otjo_df = pd.read_excel(xls, 'Otjozondjupa')
Hard_df = pd.read_excel(xls,'Hardap')
Oma_df = pd.read_excel(xls, 'Omaheke')

#You do not need to do this you can work on the df as they are above

#Concantinate dfs for 2021 data
frames = [Kho_df, Oha_df, Ero_df, Omu_df, Kun_df, Osha_df, KaE_df, KaW_df, Oshi_df, Khar_df, Zamb_df, Otjo_df, Hard_df, Oma_df]
Covid19_2020= pd.concat(frames, keys=["Khomas","Ohangwena","Erongo","Omusati","Kunene","Oshana","Kavango East","Kavango West","Oshikoto","Kharas","Zambezi",
                                       "Otjozondjupa","Hardap","Omaheke"])
#Drop unamed columnss. 
Covid19_2020 = Covid19_2020.drop(columns = ['Unnamed: 9', 'Unnamed: 10'])
```

```python
#2021 Data
#One way to import data from one excel with multiple sheets 
#Using pd.read_excel
#Raw_Khomas_2021 = pd.read_excel(r'2021_DSA_Cases.xlsx',sheet_name = 'Khomas')
#Raw_Ohangwena_2021 = pd.read_excel(r'2021_DSA_Cases.xlsx',sheet_name = 'Ohangwena')
#Continue with all the regions


## Now you can list all sheets in the file
#xls.sheet_names
#

#Reading all the sheets within the xls.
xls = pd.ExcelFile(p1)
Kho_df1 = pd.read_excel(xls, 'Khomas')
Oha_df1 = pd.read_excel(xls, 'Ohangwena')
Ero_df1 = pd.read_excel(xls, 'Erongo') 
Omu_df1 = pd.read_excel(xls, 'Omusati')
Kun_df1 = pd.read_excel(xls, 'Kunene')
Osha_df1 = pd.read_excel(xls, 'Oshana')
KaE_df1 = pd.read_excel(xls, 'Kavango East')
KaW_df1 = pd.read_excel(xls, 'Kavango West')
Oshi_df1 = pd.read_excel(xls, 'Oshikoto')
Khar_df1 = pd.read_excel(xls, 'Karas')
Zamb_df1 = pd.read_excel(xls, 'Zambezi')
Otjo_df1 = pd.read_excel(xls, 'Otjozondjupa')
Hard_df1 = pd.read_excel(xls,'Hardap')
Oma_df1= pd.read_excel(xls, 'Omaheke')

#You do not need to do this you can work on the df as they are above

#Concantinate dfs for 2021 data
frames = [Kho_df1, Oha_df1, Ero_df1, Omu_df1, Kun_df1, Osha_df1, KaE_df1, KaW_df1, Oshi_df1, Khar_df1, Zamb_df1, Otjo_df1, Hard_df1, Oma_df1]
Covid19_2021 = pd.concat(frames, keys=["Khomas","Ohangwena","Erongo","Omusati","Kunene","Oshana","Kavango East","Kavango West","Oshikoto","Kharas","Zambezi",
                                       "Otjozondjupa","Hardap","Omaheke"])
#Drop unamed columnss. 
Covid19_2021 = Covid19_2021.drop(columns = ['Unnamed: 9', 'Unnamed: 10'])
```
